```python
with open("bigquery_sql/questions_answered_distribution.sql", "r") as f:
    questions_answered_sql = f.read()

questions_answered_df = pandas_gbq.read_gbq(questions_answered_sql)
```

```python
questions_answered_df
```

```python
with plt.style.context("Solarize_Light2"):
    sns.histplot(data=questions_answered_df, x="num_questions_answered", kde=True, alpha=0.5)
    plt.legend(loc="upper right")
    # plt.xlim(0,3_000_000)
    plt.tight_layout()
    plt.title("Distribution of number of questions answered on StackOverflow", fontdict={"size":10})
    plt.show()
```
