```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
users = pd.read_csv('../data_users/user_ids.csv')
print(users.shape)
print(users.info())
print(users.head())
```

```python
# plotting counts for each year ordered by year
users['year'].value_counts().sort_index().plot(kind='bar')
plt.title('Number of users by year')
plt.show()

# by month ordered by month
# first remove any row with year==2024
user_month = users[users['year'] != 2024]
user_month['month'].value_counts().sort_index().plot(kind='bar')
plt.title('Number of users by month (excluding 2024)')
plt.show()
```
